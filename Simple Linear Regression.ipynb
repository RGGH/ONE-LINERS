```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv('Data.csv')
x = dataset.iloc[:, :-1].values # every row, up to last column
y = dataset.iloc[:, -1].values # every row , last column
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
```

```python
regression_obj = LinearRegression()
regression_obj.fit(x_train, y_train)
```

```python
from sklearn.linear_model import LinearRegression  
regression_obj = LinearRegression()
regression_obj.fit(x_train, y_train)
```

```python
y_predicted = regression_obj.predict(x_test)
```

```python
plt.scatter(x_train, y_train, color = 'red')
plt.plot(x_train, regression_obj.predict(x_train), color = 'green')
plt.title('Salary vs Experience for Training set')
plt.xlabel('Experience (in Years)')
plt.ylabel('Salary')
plt.show()
```

```python
# Visualizing the Test Set Results
plt.scatter(x_test, y_test, color='red') # Same as above, but uses the 20% test data, not 80% train data
plt.plot(x_train, regression_obj.predict(x_train), color='blue')
plt.title('Salary vs Experience for Test set')
plt.xlabel('Experience (in Years)')
plt.ylabel('Salary')
plt.show()
```
